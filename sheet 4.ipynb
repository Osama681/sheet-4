5=visualization methods to study the effect of each parameter on the reaction product

```python
df_glucose_conc= df.groupby('glucose conc [g/l]', as_index=False)['gluconic acid conc [g/l]'].sum()
df_glucose_conc
```

```python
df_glucose_conc.plot(kind='line', figsize=(15, 5), x='glucose conc [g/l]', y='gluconic acid conc [g/l]')
# plot title
plt.title('gluconic acid conc [g/l] vs glucose conc [g/l]', y=1.1)
# X-axis title
plt.xlabel('glucose conc [g/l]')
# Y-axis title
plt.ylabel('gluconic acid conc [g/l]')
# X-axis ticks/labels
plt.xticks(rotation=45)
# set X margins to 0
plt.margins(x=0)
# show grid lines
plt.grid()
```

```python
df_gluconic2 = df.groupby('biomass conc [g/l]', as_index=False)['gluconic acid conc [g/l]'].sum()
df_gluconic2
```
